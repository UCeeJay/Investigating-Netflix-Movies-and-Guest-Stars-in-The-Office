```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
import matplotlib.pyplot as plt

office_episodes = pd.read_csv("datasets/office_episodes.csv")
office_episodes
```

```python
episode_number = office_episodes['episode_number']
episode_number
```

```python
viewership_mil = office_episodes['viewership_mil']
viewership_mil
```

```python
colors = []

for i, rows in office_episodes.iterrows():
    if rows['scaled_ratings'] < 0.25:
        colors.append('red')
    elif rows['scaled_ratings'] >= 0.25 and rows['scaled_ratings'] < 0.50:
        colors.append('orange')
    elif rows['scaled_ratings'] >= 0.50 and rows['scaled_ratings'] < 0.75:
        colors.append('lightgreen')
    elif rows['scaled_ratings'] >= 0.75:
        colors.append('darkgreen')
        
colors
```

```python
scales = []

for i, rows in office_episodes.iterrows():
    if rows['has_guests'] != False:
        scales.append(250)
    else:
        scales.append(25)
        
scales
```

```python
fig = plt.figure(figsize=(11,7))

plt.scatter(x=episode_number, 
            y=viewership_mil, 
            c=colors, 
            s=scales) 

plt.title("Popularity, Quality, and Guest Appearances on the Office")
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")

plt.show()
```
